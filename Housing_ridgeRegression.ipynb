```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso

from sklearn.model_selection import train_test_split

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
train_data = pd.read_csv(data_dir + 'clean_train.csv')
test_data = pd.read_csv(data_dir + 'clean_test.csv')

Y_train = train_data.SalePrice.values
X_train = train_data.drop(['SalePrice', 'Id'], axis=1)

#Saving the Id's of the testing set to sumbit on Kaggle
ids = test_data.Id
X_test = test_data.drop(['Id'], axis=1)

X_train = X_train.values
X_test = X_test.values
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
preds = lr.predict(X_test)

submission = pd.DataFrame()

submission.loc[:, 'Id'] = ids
submission.loc[:, 'SalePrice'] = preds

submission.to_csv(data_dir + 'Kaggle_housing_preds.csv', index=False)
```

```python
rr_100 = Ridge(alpha = 3.5)
rr_100.fit(X_train, Y_train)
pred_rr = rr_100.predict(X_test)
submission = pd.DataFrame()

submission.loc[:, 'Id'] = ids
submission.loc[:, 'SalePrice'] = pred_rr

submission.to_csv(data_dir + 'Kaggle_RR9_housing_preds.csv', index=False)
```
